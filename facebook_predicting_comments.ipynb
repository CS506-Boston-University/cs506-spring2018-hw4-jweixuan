```python
from sklearn.utils import shuffle
import sklearn.model_selection as model_selection
from sklearn.model_selection import KFold


kf = KFold(n_splits = 5, shuffle = True)

mean_mse = 0


for train_index, test_index in kf.split(X):
    
    #print("TRAIN:", train_index, "TEST:", test_index)
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]
    est9 = sm.OLS(y_train,X_train)
    est9 = est9.fit()
    
    pre = est9.predict(X_test)
    mean_mse += MSE(y_test,pre)
    
#x_tra, x_te, y_tra, y_te = model_selection.train_test_split(X1, Y1, test_size=0.2, random_state=0)

print("The MSE of the linear model is :",mean_mse/5)
```

```python
kf = KFold(n_splits = 5, shuffle = True)

mean_mse = 0


for train_index, test_index in kf.split(X):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]
    est9 = sm.OLS(y_train,X_train)
    est9 = est9.fit_regularized(alpha = 0.2, L1_wt = 0)
    #print(est9.summary())
    
    pre = est9.predict(X_test)
    mean_mse += MSE(y_test,pre)
    
#x_tra, x_te, y_tra, y_te = model_selection.train_test_split(X1, Y1, test_size=0.2, random_state=0)

print("The MSE of the regularized linear model is :",mean_mse/5)
```

Try the following values for the ridge regression parameter $\alpha$ (the penalty weight), $\alpha = [0,5,10,100]$.**(1 pt)**

```python
def cal_mse_with_different_a(a):

    kf = KFold(n_splits = 5, shuffle = True)

    mean_mse = 0


    for train_index, test_index in kf.split(X):
        #print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = Y[train_index], Y[test_index]
        est9 = sm.OLS(y_train,X_train)
        est9 = est9.fit_regularized(alpha = a, L1_wt = 0)
        #print(est9.summary())

        pre = est9.predict(X_test)
        mean_mse += MSE(y_test,pre)

    #x_tra, x_te, y_tra, y_te = model_selection.train_test_split(X1, Y1, test_size=0.2, random_state=0)

    print("The MSE of the regularized linear model with alpha = ", a, " is :",mean_mse/5)
```
